```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50, 60, 70]).reshape(-1, 1)
Y = np.array([50, 100, 150, 190, 240, 280, 330])

# Fit Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Regression Line
predicted_Y = model.predict(X)

# Visualization
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, predicted_Y, color='red', label='Regression Line')
plt.xlabel('Minutes of Exercise')
plt.ylabel('Calories Burned')
plt.legend()
plt.show()
```

```python
# Predict calories burned for 45 minutes
predicted_calories = model.predict([[45]])
print(f"Calories burned for 45 minutes of exercise: {predicted_calories[0]:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
data = {
    'RAM': [4, 6, 8, 6, 8, 12],
    'Storage': [64, 128, 128, 64, 256, 256],
    'Battery': [3000, 3500, 4000, 3300, 4500, 5000],
    'Price': [250, 320, 400, 300, 450, 520]
}
df = pd.DataFrame(data)

# Model training
X = df[['RAM', 'Storage', 'Battery']]
Y = df['Price']
model = LinearRegression()
model.fit(X, Y)

# Prediction
new_phone = np.array([[8, 128, 4200]])
predicted_price = model.predict(new_phone)
print(f"Predicted price for the phone: ${predicted_price[0]:.2f}")
```

```python
print(f"RAM coefficient: {model.coef_[0]:.2f}")
print(f"Storage coefficient: {model.coef_[1]:.2f}")
print(f"Battery coefficient: {model.coef_[2]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([6, 8, 10, 12, 14, 16, 18, 20]).reshape(-1, 1)
Y = np.array([15, 17, 20, 25, 28, 26, 22, 18])

# Transform to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
X_curve = np.linspace(6, 20, 100).reshape(-1, 1)
Y_curve = model.predict(poly.transform(X_curve))

# Visualization
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_curve, Y_curve, color='red', label='Polynomial Curve')
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()

# Predict temperature for 13:00
temperature_at_13 = model.predict(poly.transform([[13]]))
print(f"Predicted temperature at 13:00: {temperature_at_13[0]:.2f}°C")
```
